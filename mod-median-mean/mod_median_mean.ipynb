<p>A sample data set was created using random.</p>

```python
salary=np.random.normal(5000,2000,1000)
salary
```

```python
#to draw
import matplotlib.pyplot as plt
plt.hist(salary,100)
plt.show()
```

```python
np.median(salary)
```

```python
#When a different value is entered
salary=np.append(salary,[100000000])
np.mean(salary)
```

```python
#median gives a more accurate value
np.median(salary)
```
